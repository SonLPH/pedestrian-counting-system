### TOP !0 (MOST PEDESTRIANS) LOCATIONS BY DAY ###

```python
day_df = pedestrian_df.groupby(["Year", "Month", "Mdate", "Day", "Sensor_ID"])["Hourly_Counts"].sum().reset_index()\
    .groupby(['Year','Month','Mdate']).apply(lambda x: x.sort_values(by=['Hourly_Counts'], ascending=False)).reset_index(drop=True)\
    .groupby((['Year','Month','Mdate'])).head(10)
```

```python
day_df
```

### TOP 10 (MOST PEDESTRIANS) LOCATIONS BY MONTH ###

```python
month_df = pedestrian_df.groupby(['Year', 'Month', 'Sensor_ID'])['Hourly_Counts'].sum().reset_index()\
    .groupby(['Year','Month',]).apply(lambda x: x.sort_values(by=['Hourly_Counts'], ascending=False)).reset_index(drop=True)\
    .groupby((['Year','Month'])).head(10)
```
